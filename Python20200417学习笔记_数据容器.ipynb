### 元组的巧妙利用!

元组容器应用举例：
def mul(x,y,z):return x\*y\*z

1. 调用方法一：
mul(10,20,30)  ==>6000

2. 调用方法二：
a,b,c=10,20,30;mul(a,b,c)==>6000

3. 调用方法三：
x=10,20,30;mul(*x)==>6000

````python
def mul(x,y,z):
    return x*y*z

print(mul(10,20,30))

a,b,c=10,20,30;
print(mul(a,b,c))

x=10,20,30;
print(mul(*x))
````

**修改函数mul的定义，可以支持任意参数个数——任意的参数列表**

````python
def mul2(*x):
    n=len(x)
    p=1
    for i in x:
         p*=i
    return p
````

````python
y=10,20,30,40
print(mul2(*y))
````

### 字典容器实例的应用
matplotlib中plot函数

##### 字典实例应用举例

````python
# coding=UTF-8

from matplotlib import pyplot as plt  #从matplotlib导入子包pyplot昵称plt
from scipy.special import jn         #从 scipy导入子包bessel函数jn

plt.rcParams['font.sans-serif']=['SimHei'] #用来正常显示中文标签
plt.rcParams['axes.unicode_minus']=False #用来正常显示负号

x=plt.np.linspace(0,30,50)  #导入子包pyplot时自动导入了数组包numpy昵称np
y=jn(0,x),jn(1,x),jn(2,x),jn(3,x) #元组容器实例y含4个成员

dt0=dict(color='r',marker='o',linestyle='-', linewidth=1.5, markersize=6)
dt1=dict(color='g',marker='s',linestyle='-.',linewidth=2,   markersize=8)
dt2=dict(color='b',marker='v',linestyle='--',linewidth=2.5, markersize=10)
dt3=dict(color='m',marker='d',linestyle=':', linewidth=3,   markersize=12)
dt=dt0,dt1,dt2,dt3  #定义0~3阶bessel函数曲线特征字典值成员组成的元组变量

for i in range(4):
    plt.plot(x,y[i],**dt[i])
    
plt.xlabel('X-轴')
plt.ylabel('Y-轴  Basselj(0:3,X)')
plt.title('贝瑟尔函数曲线')
plt.legend(['besselj(0,x)','besselj(1,x)','besselj(2,x)','besselj(3,x)'])
plt.show()
````

# numpy N维数组容器(Ndarray)
The N-dimensional array (ndarray)

ndarray是同构多维数组，它是一个元素表（通常是数字），所有类型都相同，由非负整数元组索引。在NumPy维度中称为 轴 。

一个 ndarray是具有相同类型和大小的项目的（通常是固定大小的）多维容器。 尺寸和数组中的项目的数量是由它的shape定义， 它是由N个非负整数组成的tuple（元组），用于指定每个维度的大小。 数组中项目的类型由单独的data-type object (dtype)指定， 其中一个与每个ndarray相关联。
