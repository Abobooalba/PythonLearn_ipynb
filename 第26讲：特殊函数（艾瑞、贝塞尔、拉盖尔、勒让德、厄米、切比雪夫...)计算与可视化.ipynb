## 1. 艾瑞函数airy

````python
from scipy import special
x = np.linspace(-20,1, 201)
y = special.airy(x)
plt.rcParams['axes.unicode_minus']=False
plt.plot(x,y[0],'r-.',x,y[1],'g-.',x,y[2],'b-.',x,y[3],'c-.');
plt.show()
````

````python
from scipy import special
x = np.linspace(-30, 1, 101)
ai, aip, bi, bip = special.airy(x)

import matplotlib.pyplot as plt
plt.plot(x, ai, 'r', label='Ai(x)')
plt.plot(x, bi, 'b--', label='Bi(x)')
plt.ylim(-0.5, 1.0)
plt.grid()
plt.legend(loc='upper left')
plt.show()
````

## 2. 贝瑟尔函数（柱函数，变幅度三角函数）jn(i,x),yn(i,x),kn(i,x),...

用于柱域问题

````python
from matplotlib import pyplot as plt;
from scipy import special as sp

x=plt.np.linspace(0,30,100);
y=plt.np.zeros((4,len(x)));
for i in range(4):
    y[i]=sp.jn(i,x)
    
plt.rcParams['axes.unicode_minus']=False

plt.plot(x,y[0],'r-.',x,y[1],'g-.',x,y[2],'b-.',x,y[3],'c-.',lw=4)

plt.show()
````

##  3. 拉盖尔多项式函数（用于量子力学）

python 扩展包scipy中的多项式存储的是自变量各幂次的系数，显示时默认按降幂排列各幂次的系数，但是存储时，
```
poly1d([])[0]、poly1d([])[1]、poly1d([])[2]
```
保存的是0次、1次、2次幂的系数。例如
```
laguerre(2)=poly1d([ 0.5, -2. ,  1. ])         # 对应0.5*x^2-2*x+1.
laguerre(3)=poly1d([-1/6,  1.5  , -3.,  1. ])   # 对应-x^3/6+1.5*x^2-3*x+1

laguerre(2)[0],laguere(2)[1],laguere(2)[2]
```
输出1, -2,0.5


````python
from matplotlib import pyplot as plt
from scipy import special as sp

x=plt.np.linspace(0,8,100);
la=sp.laguerre

print(la(2))
print(la(8))
y1=la(8)(x);
y2=la(10)(x);
y3=la(12)(x)

plt.plot(x,y1,'r-.',x,y2,'g-.',x,y3,'b-.',lw=3)

plt.show()
````

## 4. 勒让德多项式函数spsp.legendre(i)（用于球域问题）

legendre(2)=poly1d([ 1.5,  0. , -0.5])   #对应1.5*x^2-0.5

legendre(3)=poly1d([ 2.5,  0. , -1.5,  0. ])#对应2.5*x^3-1.5*x


````python
from matplotlib import pyplot as plt
from scipy import special as sp
x=plt.np.linspace(-1,1,100)
print(sp.legendre(8))
y1=sp.legendre(8)(x)
y2=sp.legendre(10)(x)
y3=sp.legendre(12)(x)
plt.plot(x,y1,'r-',x,y2,'g-',x,y3,'b-',lw=2)
plt.show()
````

##  5.厄米多项式函数 sp.hermite(i)（用于量子力学谐振子）
注意:
hem(3)[0]=0,hem(3)[1]=-12,hem(3)[2]=0,hem(3)[3]=8

````python
from scipy import special as sp
hem=sp.hermite
for i in range(1,5): 
    print('hermite(',i,')=',hem(i))

x=plt.np.linspace(-2,2,100)
y1=hem(1)(x)
y2=hem(3)(x)
y3=hem(5)(x)
plt.plot(x,y1,'r-',x,y2,'g-',x,y3,'b-',lw=2)
plt.show()
````

##  6. 切比雪夫多项式函数 sp.chebyc,sp.chebys(用于滤波器设计)

````python
from scipy import special as sp

cc=sp.chebyc;
cs=sp.chebys

for i in range(1,5): 
    print('chebyc(',i,')=',cc(i))
    
for i in range(1,5): 
    print('chebys(',i,')=',cs(i))

    
x=plt.np.linspace(-2,2,100)
y1=cs(1)(x)
y2=cs(3)(x)
y3=cs(5)(x)
plt.plot(x,y1,'r-',x,y2,'g-',x,y3,'b-',lw=2)
plt.show()
````
