## 创建数组容器实例的函数方法

arange,linspace,ones,zeros,diag,random.random,random.randint,mgrid,ogrid

```python
a=np.arange(1,100,7);print(a)
b=np.linspace(0,2,11);print(b)
c=np.ones((2,5));print(c)
d=np.zeros((2,6));print(d)
e=np.random.random((2,6));print(e)
f=np.random.randint(-10,20,(2,8));print(f)
g=np.diag([1,2,3,4]);print(g)
x,y=np.mgrid[-5:5:3j,-6:6:7j];
print(x);print("\n");print(y)
```

```python
c=np.ones((2,5))
d=np.zeros((2,5))
e=np.random.random((2,5))
f=np.random.randint(-10,20,(2,5))
x,y=np.ogrid[-4:4:5j,-5:5:5j]
r=np.sqrt(x**2+y**2).round(2)
print(c,d,e,f,x,y,r)
```

```python
x,y=np.ogrid[-4:4:5j,-5:5:5j] #创建列矢量x和行矢量y
r=np.sqrt(x**2+y**2).round(2)
print('x=',x); print('y=',y);print('r=',r)
```

## 将元组、列表等容器实例转换为数组实例

```python
x=(1,2,3,4)
y=[10,20,30,1]
np.array(x)
```

```python
np.array(y)
```

```python
from matplotlib import pyplot as plt
import numpy as np
x,y=np.mgrid[-12:12:600j,-12:12:600j]

x=x.astype(np.float32)
y=y.astype(np.float32)

r=np.sqrt(x**2+y**2)
z=100*np.exp(-0.5*r),250*np.cos(4*r)*np.exp(-0.5*r),50*np.sinc(r)

z[0].size
tu=np.zeros((600,600,3))
tu[:,:,0]=z[0]
tu[:,:,1]=z[1]
tu[:,:,2]=z[2]

plt.imshow(tu)
```
