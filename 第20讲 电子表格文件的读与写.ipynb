# 第20讲 电子表格文件的读与写

## 电子表格文件的读入

1、导入模块
```
import xlrd
```
2、打开Excel文件读取数据
```
data = xlrd.open_workbook('excelFile.xls')
```
3、使用技巧——获取一个工作表
```
table = data.sheets()[0] #通过索引顺序获取
table = data.sheet_by_index(0) #通过索引顺序获取
table = data.sheet_by_name(u'Sheet1')#通过名称获取
```
4、逐行或逐列打印表内容
```
nrow=table.nrows;ncol=table.cols
for i in range(nrow):print(table.row_values(i))  #逐行打印
for j in range(ncol):print(table.col_values(j))  #逐列打印

for i in range(nrow):                  #逐cell打印
    for j in range(ncol):
        print(table.cell_value(i,j),end=",")
    print("")
```    
    
5、逐行逐列显示表的内容及数据类型

0表示空字符，1表示text,2表示number,...
```
nrow=table.nrows;
ncol=table.cols

for i in range(nrow):print(table.row(i))  #逐行打印

for j in range(ncol):print(table.col(j))  #逐列打印

for i in range(nrow):   #逐cell打印
    for j in range(ncol):
        print(table.cell(i,j),end=",")
    print("")
```

## 电子表格读取示例

````python
import xlrd as xr
data=xr.open_workbook('excelindemo.xlsx')
t=data.sheets()[0],data.sheets()[1];

nr=t[0].nrows,t[1].nrows;
nc=t[0].ncols,t[1].ncols

for k in range(2):
    for i in range(nr[k]):
        print(t[k].row_values(i))
        
for k in range(2):
    for i in range(nc[k]):
        print(t[k].col_values(i))
        
for k in range(2):
    for i in range(nr[k]):
        for j in range(nc[k]):
             print(t[k].cell_value(i,j),end=',')
        print('\n')
        
for k in range(2):
    for i in range(nr[k]):
        print(t[k].row(i))
        
for k in range(2):
    for i in range(nc[k]):
        print(t[k].col(i))
        
for k in range(2):
    for i in range(nr[k]):
        for j in range(nc[k]):
              print(t[k].cell(i,j),end=',')
        print('\n')
        
````

## 电子表格文件的创建

1、导入模块
```
import xlwt,xlrd
```
2、在内存中创建Excel工作簿
```
wb = xlwt.Workbook(encoding='utf-8')
```
3、在工作簿中创建工作表
```
sheet1=wb.add_sheet('mysheet1',cell_overwrite_ok=True) 
```
4、准备待写的二维数据列表data,例如
```
data=[['序号','姓名','部门','学位','专业','年龄','工资','sci论文'],
[1,'李玉玲','化学','硕士','无机',31,4500,2],
[2,'王树伟','物理','博士','基本粒子',29,5200,12],
[3,'曾翔宇','生物','学士','植物',40,4800,5],
[4,'刘柳清','机械','博士','机电一体化',33,5800,9]]
```
5. 将数据写入电子表格nr=len(data);nc=len(data[0])

```
for i in range(nr):
      for j in range(nc):      
           sheet1.write(i,j,data[i][j])
```

6. 将表格数据存入磁盘文件
```
wb.save('myexcel.xls');
```

````python
# coding = utf-8
import xlrd; 
import xlwt;
import datetime

#workbook = xlwt.Workbook(encoding='UTF-8') #用xlwt 创建工作簿， encoding是编码格式

workbook = xlwt.Workbook(encoding='GB2312') #用xlwt 创建工作簿， encoding是编码格式

# 添加一个sheet
sheet1 = workbook.add_sheet('mysheet1', cell_overwrite_ok=True)#在工作簿中创建工作表

# 向sheet中写入数据列表

data=[['姓名', '年龄', '收入', '党员', '职业'], 
      ['李明', 65, 7000, True, '教育'], 
      ['刘霖', 56, 5000, False, '医生'], 
      ['张辉', 52, 3800, False, '装修师']]

dn=len(data);dc=len(data[0])

for i in range(dn):
    for j in range(dc):
        sheet1.write(i,j,data[i][j]) #写数据到工资表的单元格
    
workbook.save('exceloutdemo.xlsx') #保存工作簿

wb=xlrd.open_workbook('exceloutdemo.xlsx') #用xlrd打开工作簿

sh1=wb.sheets()[0];
nr=sh1.nrows
for i in range(nr):
    print(sh1.row_values(i))
````
