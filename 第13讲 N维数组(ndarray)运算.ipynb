## 数组的排序

```python
import numpy as np; a=np.random.randint(10,100,20);
print(a);
a.sort(kind='heapsort');print(a)
a=np.array([[1,2,3],[40,50,60],[7,-8,90]]);print(a);
a.sort(axis=0);print(a)
a.sort(axis=1);print(a)
```

## 数组排序的可视化观感
调用numpy随机模块random的函数生成400*600个随机数，将数组和三种排序结果可视化为四幅图片，体验排序的直觉效果。

```python
from matplotlib import pyplot as pt
import numpy as np

a=np.random.random((400,600)).astype(pt.np.float32).round(3)

a0=np.sort(a,axis=0) #竖排序
a1=np.sort(a,axis=1) #横排序
a2=np.sort(a,axis=None).reshape(400,600) #展平排序再折叠
tu=a,a0,a1,a2  #tuple数组（每个成员是数组的元组容器实例)
c='spring','summer','autumn','winter'  # string元组（每个成员是 字符的元组容器实例）
# pc=dict(num=0, figsize=(200,300), dpi=20, edgecolor='green', facecolor=None)
# pt.figure(**pc)  #创建画图窗口
for i in range(4): 
    pt.subplot(2,2,i+1)
    pt.imshow(tu[i],cmap=c[i])
    pt.colorbar()
pt.show()
print("我是一位粉刷匠，乱点鸳鸯，竖着刷，横着刷，最后这幅图又咋刷？")
```

## 数组部分和整体和、累加和

```python
import numpy as np; 
ar=np.arange(1,21).reshape(4,5);print(ar)
asc=ar.sum(axis=0);print('列和=',asc)
asr=ar.sum(axis=1);print('行和=',asr)
asum=ar.sum();print('整体和=',asum)
print('累加和=',ar.cumsum())
ar.cumsum(axis=0) #列累加和
```

```python
import numpy as np; 
a=np.ones((1,10))
print(a)
a1=np.cumsum(a)
print('累加和',a1)     
a=np.arange(1,11).reshape(2,5);
print(a)
asc=a.sum(axis=0);print('列和=',asc)
asr=a.sum(axis=1);print('行和=',asr)
asum=a.sum();print('整体和=',asum)
print('累加和=',a.cumsum())
```

## numpy数组累乘——用于求阶乘

```python
import numpy as np;
ar=np.arange(1,6);
print(ar,ar.cumprod())
ar=np.arange(1,13).reshape(3,4);
print(ar)
print(ar.cumprod(axis=0))#列向累乘
print(ar.cumprod(axis=1))#行向累乘
```

## 数组极值和极值索引

```python
import numpy as np;
ar=np.array([[11,78,-90],[-23,56,10],[-2,67,98]]);
print(ar.max(axis=0),ar.max(axis=1),ar.max())
print(ar.min(axis=0),ar.min(axis=1),ar.min())
print(ar.argmax(axis=0),ar.argmax(axis=1),ar.argmax())
print(ar.argmin(axis=0),ar.argmin(axis=1),ar.argmin())
```
