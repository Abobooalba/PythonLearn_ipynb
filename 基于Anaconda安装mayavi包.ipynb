````python
import numpy as np
from mayavi import mlab as mb
x,y,z=np.mgrid[-10:10:40j,-10:10:40j,-6:6:25j];
s=(2*x**2+3*y**2-0.5*z**2)*np.exp(-0.125*(x**2+y**2+z**2))
cm1="summer";
cm2="jet"
mb.points3d(x,y,z,s,colormap=cm1)
mb.show()
````
